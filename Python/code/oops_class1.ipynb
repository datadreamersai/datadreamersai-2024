```python
# Object-Oriented Programming (OOP) Approach
class AtmFunctions:#ATM functionality
    def __init__(self, account_number, balance=0):
        self.account_number = account_number  # Each account has a unique ID
        self.balance = balance  # Each account has its own balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount} into account {self.account_number}")
        else:
            print("Invalid deposit amount!")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn: {amount} from account {self.account_number}")
        else:
            print("Insufficient balance or invalid amount!")

    def check_balance(self):
        print(f"Account {self.account_number} Balance: {self.balance}")



class LoanFunction():
  def __init__(self, interest_rate=0.1):
    self.interest_rate = interest_rate

  def intrest_calculation(self,principle,time):
    s_i = (principle*self.interest_rate*time)/100
    return "Your intrest is "+str(s_i)


  def floreclosure(self,no_of_months_paid):
    if no_of_months_paid>12:
      return "You are eligible for foreclosure"
    else:
      return "You are not eligible for foreclosure"




# Creating multiple accounts dynamically
account1 = AtmFunctions("A123", 1000)
account2 = AtmFunctions("B456", 500)

# Using the same functions without separate balance variables
account1.deposit(500)
account1.withdraw(200)
account1.check_balance()

account2.deposit(300)
account2.withdraw(100)
account2.check_balance()



lf = LoanFunction()
print(lf.intrest_calculation(50000,12))
print(lf.floreclosure(10))

```

```python
class AtmMachineFunctions():
  def __init__(self):
    self.pin = input("Please enter your pin: ")
    self.balance = 10000
    self.menu()

  def menu(self):
    option = input('''
    1.check_balance
    2.deposit
    3.withdraw
    4.change pin
    5.loan credibility check
    ''')
    if option == '1':
      self.check_balance()
      self.menu()

    elif option == '2':
      amount = int(input("Enter amount to deposit: "))
      self.deposit(amount)
      self.menu()

    elif option == '3':
      amount = int(input("Enter amount to withdraw: "))
      self.withdraw(amount)
      self.menu()

    elif option == '4':
      old_pin = input("Enter your old pin: ")
      self.change_pin(old_pin)
      self.menu()

    elif option =='5':
      amount = input("Enter the loan amount: ")
      tenure = input("For how many months do u need loan : ")
      self.loan_credibility_check(amount,tenure)
      self.menu()

    else:
      print("Invalid option")
      exit()




  def check_balance(self):
    print("Your balance is : {}".format(self.balance))

  def deposit(self,amount):
    self.balance = self.balance + amount
    print("Deposit is successfull","Your balance is : {}".format(self.balance))


  def withdraw(self,amount):
    if amount>self.balance:
      print("Insufficient balance")
    else:
      self.balance = self.balance - amount
      print("Withdraw is successfull","Your balance is : {}".format(self.balance))



  def change_pin(self,old_pin):
    if old_pin == self.pin:
      new_pin = input("Enter new pin: ")
      self.pin = new_pin
      print("Pin changed successfully")
    else:
      print("Wrong pin")

  def loan_credibility_check(self,amount,tenure):
    if int(amount)<100000 and int(tenure) <60:
      print("You are eligible for loan")
    else:
      print("You are not eligible for loan")

```

**magic methods**

```python
class Point():

  def __init__(self,a,b):
    self.x_cord = a
    self.y_cord = b

  def __str__(self):
    return "({},{})".format(self.x_cord,self.y_cord)


  def __add__(self,other):
    x = self.x_cord + other.x_cord
    y = self.y_cord + other.y_cord
    return "({},{})".format(x,y)


  def __sub__(self,other):
    x = self.x_cord - other.x_cord
    y = self.y_cord - other.y_cord
    return "({},{})".format(x,y)


  def __mul__(self,other):
    x = self.x_cord * other.x_cord
    y = self.y_cord * other.y_cord
    return "({},{})".format(x,y)


  def __truediv__(self,other):
    x = self.x_cord / other.x_cord
    y = self.y_cord / other.y_cord
    return "({},{})".format(x,y)

```

```python
p1 = Point(2,3)
p2=Point(4,5)
```
