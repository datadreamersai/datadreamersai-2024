```python
def func1(val):

  if pd.notna(val):


    return "".join(val.split(","))
  else:
    return val

```

```python
df.Gross = df.Gross.apply(func1)
```

```python
df.Gross = df.Gross.apply(lambda x : int(x) if pd.notna(x) else x)
```

```python
top_10_movies = df[["Series_Title","Meta_score","Genre"]].sort_values(by="Meta_score",ascending=False).iloc[0:10]
```

```python
top_10_movies
```

```python
top_10_movies.Genre.to_list()
```

```python
list_of_genre =[]
for ele in top_10_movies.Genre.to_list():
  list_of_genre.append(ele.split(", "))
```

```python
list_of_genre
```

```python
de_nested_genre=[]
def de_nesting_genre(ele):
  for i in ele:
    de_nested_genre.append(i)

for ele in list_of_genre:
  de_nesting_genre(ele)
```

```python
de_nested_genre
```

```python
genre_count = {}

for ele in de_nested_genre:
  if ele in genre_count:
    genre_count[ele]=genre_count.get(ele)+1
  else:
    genre_count[ele]=1
```
