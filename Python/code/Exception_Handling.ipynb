```python
#raise keyword

def withdraw(amount,balance=10000):
  if amount>balance:
    raise Exception("Insufficient balance")
  elif amount<0:
    raise Exception("Amount cannot be negative")
  elif amount==0:
    raise Exception("Amount cannot be zero")
  elif amount/0:
    raise ZeroDivisionError("Cannot divide with zero")
  else:
    withdraw_amount = balance-amount
  return withdraw_amount


try:
  withdraw_amount=withdraw(1000)
except ZeroDivisionError as e:
  print(e)
except Exception as e:
  print(e)
else:
  print("Please collect your withdraw amount {}.".format(withdraw_amount))

```
