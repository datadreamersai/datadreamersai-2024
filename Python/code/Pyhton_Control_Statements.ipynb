```python

#multiple if else

x = int(input("Please enter your marks : "))

if x>=900:
  print("You are eligible for tier1 colleges")
elif x>=800 and x<900:
  print("You are eligible for tier2 colleges")
elif x>=500 and x<800:
  print("You are eligible for tier3 colleges")
else:
  print("You are eligible for tier4 colleges")
```

```python
#nested if else

marks = int(input("Enter your marks : "))
age = int(input("Enter your age : "))

if marks>=900:
  if (age>=18) and (age<20):
    print("You are eligible for tier1 premium colleges")
  else:
    print("You are eligible for tier1 regular colleges")

elif marks>=800 and marks<900:
  if age>=20 and age<25:
    print("You are eligible for tier2 premium colleges")
  else:
    print("You are eligible for tier2 regular colleges")

elif marks>=500 and marks<800:
  print("You are eligible for tier3 colleges")
else:
  print("You are eligible for tier4 colleges")
```
