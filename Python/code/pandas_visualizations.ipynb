```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data
data = {
    'CustomerID': ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10'],
    'Tenure': [5, 20, 10, 30, 50, 40, 15, 25, 35, 60],
    'MonthlyCharges': [70, 45, 65, 80, 100, 90, 55, 75, 85, 110],
    'TotalCharges': [350, 900, 650, 2400, 5000, 3600, 825, 1875, 2975, 6600],
    'Churn': ['Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)
df
```

**Box Plot**

```python
churn_targets = df.Churn.value_counts().to_dict().keys()
targets_count = df.Churn.value_counts().to_dict().values()
```

**Churn Count**

```python
#2 bar graphs
plt.figure(figsize=(5,5))
plt.bar(churn_targets, targets_count,color=["blue","red"],label=["no","yes"],width=0.5,edgecolor="black")
plt.xlabel('Churn Targets')
plt.ylabel('Count')
plt.title('Churn Targets Count')
plt.legend(title="Churn Targets")
plt.show()
```

**Customer Tenure**

```python
#multi bar graphs
plt.figure(figsize=(10,10))
plt.bar(df.CustomerID,df.Tenure,color=["red","green","blue","orange","yellow","red","green","blue","orange","yellow"],label=["C1","C2","C3","C4","C5","C6","C7","C8","C9","C10"],edgecolor="black")
plt.legend(title="Customer ID")
plt.xlabel('Customer ID')
plt.ylabel('Tenure')
plt.title('Customer Tenure')
plt.show()
```

**Churn Count using seaborn**

```python
plt.figure(figsize=(6, 4))
ax = sns.barplot(x="Churn",y="Tenure",hue="Churn",data=df,palette="coolwarm")

# Add labels using bar_label (Matplotlib 3.4+)
for container in ax.containers:
    ax.bar_label(container, fontsize=10,fontweight='bold')
plt.title("Churn Count", fontsize=14, fontweight="bold")
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Count", fontsize=12)
```

**Customer Tenure using seaborn**

```python
plt.figure(figsize=(6, 4))
ax = sns.barplot(x="CustomerID",y="Tenure",hue="Churn",data=df,palette="deep")

# Add labels using bar_label (Matplotlib 3.4+)
for container in ax.containers:
    ax.bar_label(container, fontsize=10, fontweight='bold')
plt.title("Per-customer_Tenure", fontsize=14, fontweight="bold")
plt.xlabel("CustomerID", fontsize=12)
plt.ylabel("Tenure_in_months", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.9)
```

**Scatter Plot - Tenure vs Monthly Charges**

```python
#matplotlib
plt.figure(figsize=(6, 4))
plt.scatter(df['Tenure'], df['MonthlyCharges'], color='blue', alpha=0.3)#alpha talks about opacity(transparent)
plt.title("Tenure vs Monthly Charges", fontsize=14, fontweight="bold")
plt.xlabel("Tenure (months)", fontsize=12)
plt.ylabel("Monthly Charges ($)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.9)
plt.show()
```

```python
#seaborn
plt.figure(figsize=(6, 4))
sns.scatterplot(x="Tenure", y="MonthlyCharges", data=df, hue="Churn", style="Churn",palette="deep",edgecolor="black")
plt.title("Tenure vs Monthly Charges", fontsize=14, fontweight="bold")
plt.xlabel("Tenure (Months)", fontsize=12)
plt.ylabel("Monthly Charges ($)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**Boxplot - Monthly Charges by Churn**

```python
import matplotlib.pyplot as plt

# Splitting the data based on Churn values
churn_yes = df[df['Churn'] == 'Yes']['MonthlyCharges']
churn_no = df[df['Churn'] == 'No']['MonthlyCharges']

print(churn_yes)
print(churn_no)

# Creating the boxplot
plt.figure(figsize=(6, 4))
plt.boxplot([churn_no, churn_yes], tick_labels=['No', 'Yes'],patch_artist=True, boxprops=dict(facecolor='lightblue'),label=["no","yes"])#patch_artist=True==>(Enables filling the box with color),boxprops=dict(facecolor='lightblue')==>	Sets the box fill color to light blue

# Adding titles and labels
plt.title("Monthly Charges by Churn", fontsize=14, fontweight="bold")
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Monthly Charges ($)", fontsize=12)
plt.legend(title="Churn")

# Displaying the plot
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
#seaborn
plt.figure(figsize=(6, 4))
sns.boxplot(x="Churn", y="MonthlyCharges", hue="Churn",data=df, palette="coolwarm")
plt.title("Monthly Charges by Churn", fontsize=14, fontweight="bold")
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Monthly Charges ($)", fontsize=12)
plt.legend(("yes","no"))
plt.grid(True,linestyle="--",alpha=0.5)
plt.show()
```

**Histogram - Tenure Distribution**

```python
plt.figure(figsize=(6, 4))
plt.hist(df["Tenure"], bins=10, color="blue", edgecolor="black", alpha=0.7)
plt.title("Tenure Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Tenure (Months)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

**Heatmap - Correlation Matrix**

```python
import numpy as np

plt.figure(figsize=(6, 4))
corr = df.drop(columns=["CustomerID", "Churn"]).corr()
cmap = plt.get_cmap("coolwarm")
plt.imshow(corr, cmap=cmap, interpolation='nearest')
plt.colorbar()
plt.xticks(np.arange(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(np.arange(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap", fontsize=14, fontweight="bold")
plt.show()
```

```python
#seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14, fontweight="bold")
plt.xticks(np.arange(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(np.arange(len(corr.columns)), corr.columns, rotation=45)
plt.show()
```

**violinplot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Assume df is your DataFrame with "MonthlyCharges" and "Churn" columns

# Separate data for Churn categories
churn_yes = df[df["Churn"] == "Yes"]["MonthlyCharges"]
churn_no = df[df["Churn"] == "No"]["MonthlyCharges"]

# Create violin plot
plt.figure(figsize=(6, 4))
plt.violinplot([churn_no, churn_yes], showmedians=True)

# Customizing plot
plt.xticks([1, 2], ["No", "Yes"])  # X-axis labels
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Monthly Charges ($)", fontsize=12)
plt.title("Monthly Charges by Churn (Violin Plot)", fontsize=14, fontweight="bold")

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.violinplot(x="Churn", y="MonthlyCharges", hue="Churn",data=df, palette="coolwarm", inner="box")

# Customizing plot
plt.title("Monthly Charges by Churn (Violin Plot)", fontsize=14, fontweight="bold")
plt.xlabel("Churn", fontsize=12)
plt.ylabel("Monthly Charges ($)", fontsize=12)

# Show plot
plt.show()
```

**pie chart**

```python
import matplotlib.pyplot as plt

# Count the number of customers in each Churn category
churn_counts = df["Churn"].value_counts()

# Labels and sizes
labels = churn_counts.index  # "Yes" and "No"
sizes = churn_counts.values  # Count of customers
colors = ["lightblue", "salmon"]  # Custom colors

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors,  startangle=140,wedgeprops={'edgecolor': 'black'})

# Customizing plot
plt.title("Customer Churn Distribution", fontsize=14, fontweight="bold")

# Show plot
plt.show()
```
